```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = np.array([
[57,41,41,49,51],
[61,66,70,62,65],
[187,195,191,182,177],
[207,224,213,220,230],
[300,301,290,304,300],
])

column_names = ['2007','2008','2009','2010','2011']
row_names = ['Assault','Cerebrovescular','Suicide','Accidents','Diabetes']

fig = plt.figure()
fig = pl.figure(figsize=(13,13))
ax = Axes3D(fig)
plt.title('3 D plot of Cause of death of White Male in NYC from 2007 to 2011)', fontsize=20)
lx= len(data[0])          # Work out matrix dimensions
ly= len(data[:,0])
xpos = np.arange(0,lx,1)    # Set up a mesh of positions
ypos = np.arange(0,ly,1)
xpos, ypos = np.meshgrid(xpos+0.25, ypos+0.25)

xpos = xpos.flatten()   # Convert positions to 1D array
ypos = ypos.flatten()
zpos = np.zeros(lx*ly)

dx = 0.5 * np.ones_like(zpos)
dy = dx.copy()
dz = data.flatten()
ax.set_color_cycle(kelly_colors_hex)  ## choosing from the color blind 
cs = [kelly_colors_hex[19],kelly_colors_hex[8],kelly_colors_hex[2],kelly_colors_hex[0],kelly_colors_hex[6]] * ly

ax.bar3d(xpos,ypos,zpos, dx, dy, dz, color=cs,alpha=0.5)

#sh()
ax.w_xaxis.set_ticklabels(column_names)
ax.w_yaxis.set_ticklabels(row_names)
ax.set_xlabel('Year',fontsize=15)
ax.set_ylabel('Cause Of Death',fontsize=15)
ax.set_zlabel('Number of Deaths',fontsize=15)
pl.savefig('causeOFDeath.png')
plt.show()
```
