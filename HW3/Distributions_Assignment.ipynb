##3 
PLOT the distributions of all sample means (together for all distributions). 
Mandatory  - plot is as a histogram, 
Optional - plot it  in any other way you think is convincing



```python
sampleSize = []
sample = {}
mean_sample={}
sample['normalDistribution']={}
sample['ChiSquare']={}
sample['Poisson']={}
sample['Binomial'] ={}
sample['Rayleigh'] = {}
sample['Triangular'] = {}
mean_sample['normalDistribution']={}
sampleSize= (np.random.randint(10,2000,100))
mean_sample_ND=[]
mean_sample_CS=[]
mean_sample_PO=[]
mean_sample_BI=[]
mean_sample_RA=[]
mean_sample_TR=[]

for s in sampleSize:
    sample['normalDistribution'][s]=np.random.normal(200,1,s)
    sample['ChiSquare'][s]=np.random.chisquare(200,s)
    sample['Poisson'][s]  = np.random.poisson(200,s)
    sample['Binomial'][s] =np.random.binomial(400,0.5,s)
    sample['Rayleigh'][s] =np.random.rayleigh(159.576,s)
    sample['Triangular'][s]= np.random.triangular(100,200,300,s)
    for m in sample['normalDistribution'].iterkeys():
        #mean_sample['normalDistribution'][m]=sample['normalDistribution'][m].mean()
        mean_sample_ND.append(np.random.normal(200,1,s).mean())
    for m in sample['ChiSquare'].iterkeys():
        mean_sample_CS.append(np.random.chisquare(200,m).mean())
    for m in sample['Poisson'].iterkeys():
        mean_sample_PO.append(np.random.poisson(200,m).mean())
    for m in sample['Binomial'].iterkeys():
        mean_sample_BI.append(np.random.binomial(400,0.5,m).mean())
    for m in sample['Rayleigh'].iterkeys():
        mean_sample_RA.append(np.random.rayleigh(159.576,m).mean())
    for m in sample['Triangular'].iterkeys():
        mean_sample_TR.append(np.random.triangular(100,200,300,m).mean())

plt.figure(figsize=(10,10))
plt.hist(mean_sample_ND,label='Normal')
plt.hist(mean_sample_CS,alpha=0.3,label='Chisquare')
plt.hist(mean_sample_PO,label = 'Poisson')
plt.hist(mean_sample_BI,label = 'Binomial')
plt.hist(mean_sample_RA,alpha = 0.1,label= 'Rayleigh')
plt.hist(mean_sample_TR,alpha = 0.4,label= 'Triangular')
plt.title('Plot of all the distributions mean')
plt.xlabel('sample mean')
plt.legend()
axvline(200,color='yellow',linewidth=(1),linestyle='--')

```
